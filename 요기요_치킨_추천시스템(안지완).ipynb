matrix factorialization

```python
df_user_food_ratings = df.pivot_table(
    index = '사용자id',
    columns = '가게',
    values = '총점',
    aggfunc = np.mean
).fillna(0)

df_user_food_ratings.head()
```

```python
matrix = df_user_food_ratings.values
user_ratings_mean = np.mean(matrix, axis = 1)
matrix_user_mean = matrix - user_ratings_mean.reshape(-1,1)
```

```python
matrix
```

```python
pd.DataFrame(matrix_user_mean, columns = df_user_food_ratings.columns).head()
```

```python
from scipy.sparse.linalg import svds

U, sigma, Vt = svds(matrix_user_mean, k = 16)
```

```python
print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

```python
sigma = np.diag(sigma)
```

```python
sigma.shape
```

```python
sigma[0]
```

```python
svd_user_predicted_ratings = np.dot(np.dot(U,sigma),Vt) + user_ratings_mean.reshape(-1,1)
```

```python
df_svd_preds = pd.DataFrame(svd_user_predicted_ratings, columns = df_user_food_ratings.columns)
df_svd_preds.head()
```

```python
def recommend_chickens(predictions, user_id, chickens_df, ratings_df, num_recommendations=5):
    user_idx = user_id - 1 
    sorted_user_predictions = predictions.iloc[user_idx].sort_values(ascending=False)
    user_data = ratings_df[ratings_df.user_id == user_id]

    user_history = user_data.merge(chickens_df, on = '가게').sort_values(['총점'], ascending=False)

    recommendations = chickens_df[~chickens_df['가게'].isin(user_history['가게'])]
    recommendations = recommendations.merge(pd.DataFrame(sorted_user_predictions).reset_index(), on = '가게')

    recommendations = recommendations.rename(columns = {user_idx: 'Predictions'}).sort_values('Predictions', ascending = False).iloc[:num_recommendations, :]
    return user_history, recommendations
```
