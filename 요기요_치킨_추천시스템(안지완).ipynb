matrix factorialization

```python
df_user_food_ratings = df.pivot_table(
    index = '사용자id',
    columns = '가게',
    values = '총점',
    aggfunc = np.mean
).fillna(0)

df_user_food_ratings.head()
```

```python
matrix = df_user_food_ratings.values
user_ratings_mean = np.mean(matrix, axis = 1)
matrix_user_mean = matrix - user_ratings_mean.reshape(-1,1)
```

```python
matrix
```

```python
pd.DataFrame(matrix_user_mean, columns = df_user_food_ratings.columns).head()
```

```python
from scipy.sparse.linalg import svds

U, sigma, Vt = svds(matrix_user_mean, k = 16)
```

```python
print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

```python
sigma = np.diag(sigma)
```

```python
sigma.shape
```

```python
sigma[0]
```

```python
svd_user_predicted_ratings = np.dot(np.dot(U,sigma),Vt) + user_ratings_mean.reshape(-1,1)
```

```python
df_svd_preds = pd.DataFrame(svd_user_predicted_ratings, columns = df_user_food_ratings.columns, index = df_user_food_ratings.index)
df_svd_preds
```

```python
import time

def make_wordlist(reviews,stopwords): #위의 과정들을 함수로 만든다. 이 함수는 리뷰와 종결어미 등의 stopwords를 제공하면 유의미한 단어들이 담긴 리스트를 반환한다다.
  print("리뷰 정제 중...")
  words = extract_word(reviews)
  print("형태소 추출 중...")
  words = okt.morphs(words,stem=True) 
  print("한글자 제거 중...")

  words = [x for x in words if len(x)>1 or x=="닭" or x == "맛"]
  print("불용어 제거 중...")

  words = [x for x in remove_one_word if x not in stopwords]
  print("의미있는 단어리스트 생성 중...")

  time.sleep(1)
  minimum_count = 10
  final = []
  for i in range(len(words)):
      tmp = words[i]
      if words.count(tmp) >= minimum_count:
          final.append(tmp)
  return set(final)
```

```python
def preprocess(text, word_list): #이 함수는 형태소 단위로 문장을 전처리하는 함수다.
    text = extract_word(text)
    okt = Okt()
    text = okt.morphs(text, stem = True)
    text = [x for x in text if x in word_list]
    return text
```

```python
word_list = make_wordlist(" ".join(df["리뷰"].tolist()), stopwords) #데이터프레임의 리뷰 부분에서 유의미한 단어들을 추출한다.
```
